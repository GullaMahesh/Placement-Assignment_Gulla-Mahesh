```python
import requests
import pandas as pd
```

```python
# download the dataset
url = "https://data.nasa.gov/resource/y77d-th95.json"
response = requests.get(url)#retrives the content
data = response.json()# we can access json data using json()

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)
```

```python
df.head()
```

- Get all the Earth meteorites that fell before the year 2000

```python
df["year"].isnull().sum()
```

```python
# removel of null values in year
df=df[df.year.isna()==False]
```

```python
df["year"].isnull().sum()
```

```python
# splitting the year column and retreiving the year
def getyear(year):
    return year.split("-")[0]
```

```python
df["New_year"] = df["year"].apply(getyear)
```

```python
df["New_year"] = df["New_year"].astype(int)
```

```python
type(df["New_year"][0])
```

```python
df.head()
```

```python
meteorites=df[df["New_year"]<2000]["name"]
```

```python
list(meteorites)
```

- Get all the earth meteorites co-ordinates who fell before the year 1970

```python
df["geolocation"].isnull().sum()
```

```python
# removing null values in geolocation
df=df[df.geolocation.isna()==False]
```

```python
df["geolocation"].isnull().sum()
```

```python
df["geolocation"][0]['coordinates']
```

```python
# return the cordinates
def get_cordinate(geolocationn):
    return geolocationn['coordinates']
```

```python
df["coordinates"] = df["geolocation"].apply(get_cordinate)
```

```python
df["coordinates"]
```

```python
coordinates=df[df["New_year"]<1970]["coordinates"]
```

```python
list(coordinates)
```
